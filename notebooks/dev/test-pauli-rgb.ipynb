```python
%load_ext autoreload
%autoreload 2
from polsarpro.io import open_netcdf_beam
from polsarpro.util import pauli_rgb, S_to_T3, S_to_C3, boxcar, multilook
from pathlib import Path
```

```python
input_data = Path("/data/psp/test_files/SAN_FRANCISCO_ALOS1_slc.nc")
S = open_netcdf_beam(input_data)
```

```python
rgb = pauli_rgb(S_to_C3(S), q=0.98)
rgb[:,::8].plot.imshow(interpolation="bilinear")
```

```python
C3 = S_to_C3(S)
C3 = multilook(C3, dim_az=4, dim_rg=1)
C3 = boxcar(C3, dim_az=3, dim_rg=3)
rgb = pauli_rgb(C3)
rgb[:,::2].plot.imshow()
```
