```python
from polsarpro.io import open_netcdf_beam
from polsarpro.speckle_filters import refined_lee
from polsarpro.decompositions import freeman
from polsarpro.util import S_to_C3, multilook
```

```python
S = open_netcdf_beam("/data/psp/biomass_test.nc")
```

```python
S
```

```python
C3 = S_to_C3(S)
```

```python
C3 = multilook(C3, dim_az=6, dim_rg=1)
```

```python
out = freeman(C3, boxcar_size=[5, 5])
```

```python
from dask.diagnostics import ProgressBar
# execution happens now!
with ProgressBar():
    out.to_netcdf("/data/psp/test.nc")
```

```python
import xarray as xr
out = xr.open_dataset("/data/psp/test.nc")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.imshow(out.odd, interpolation="none", vmax=1)
plt.colorbar(fraction=0.046, pad=0.04)
```
